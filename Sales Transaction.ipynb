```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sales Transaction.csv")
#
df['Date'] = pd.to_datetime(df['Date'])
#
df = df[df['TransactionNo'].str[0]!='C']
```

```python
df['Year'] = df['Date'].dt.year
#
df['Month'] = df['Date'].dt.month
#
df['Day'] = df['Date'].dt.day_name()
#
df
```

```python
df = df[df['Year'] == 2019]
df
```

```python
df['CurrentMonth'] = pd.to_datetime(df['Date']).dt.year.map(str) + "-" + pd.to_datetime(df['Date']).dt.month.map(str) + "-" '01'
#
df['CurrentMonth'] = pd.to_datetime(df['CurrentMonth']).dt.date
```

```python
df['CohortMonth'] = df.groupby('CustomerNo')['CurrentMonth'].transform('min')
#
df['CohortMonth'] = pd.to_datetime(df['CohortMonth'])
```

```python
# Code. Step 02:
def MonthDiff (df,col1,col2):
    Current_Month = pd.to_datetime(df[col1]).dt.month
    Cohort_Month = pd.to_datetime(df[col2]).dt.month
    return Current_Month - Cohort_Month 

df['MonthDiff'] = MonthDiff(df,'CurrentMonth','CohortMonth')
```

```python
# Code. Step 03:
def YearDiff (df,col1,col2):
    Current_Year = pd.to_datetime(df[col1]).dt.year
    Cohort_Year = pd.to_datetime(df[col2]).dt.year
    return Current_Year - Cohort_Year

df['YearDiff'] = YearDiff(df,'CurrentMonth','CohortMonth')
```

```python
# Code. Step 04:
def CohortIndex (df,col1,col2):
    return df[col1] * 12 + df[col2]

df['CohortIndex'] = CohortIndex(df,'YearDiff','MonthDiff')
df
```

```python
df['TotalRevenue'] = df['Price'] * df['Quantity']
```

# Products Performance Condition

```python
TV = df.groupby('ProductName')[['TransactionNo']].nunique()
RV = df.groupby('ProductName')[['TotalRevenue']].sum()
```

```python
TV['TransactionNo'].quantile([0.25,0.75])
```

```python
RV['TotalRevenue'].quantile([0.25,0.75])
```

```python
TV['TV_Score'] = pd.cut(TV['TransactionNo'],bins=[0,18,163,np.Inf],labels=['Low','Normal','High'])
TV
```

```python
RV['RV_Score'] = pd.cut(RV['TotalRevenue'],bins=[0,880,16802,np.Inf],labels=['Low','Normal','High'])
RV
```

```python
TV = TV.drop(columns=['TransactionNo'])
RV = RV.drop(columns=['TotalRevenue'])
```

```python
df = df.merge(TV,on='ProductName',how='left')
```

```python
df = df.merge(RV,on='ProductName',how='left')
```

```python
df['ProductScore'] = df.apply(lambda row:f"{row['TV_Score']}_{row['RV_Score']}",axis=1)
df
```

```python
df['ProductScore'] = df['ProductScore'].map({
    'High_High':'VeryPopular','High_Normal':'Popular','High_Low':'Normal',
    'Normal_High':'Popular','Normal_Normal':'Normal','Normal_Low':'Normal',
    'Low_High':'Normal','Low_Normal':'Normal','Low_Low':'Low'})
df
```

```python
ProductTrend = df.groupby('ProductScore').agg({'ProductName':'nunique','TransactionNo':'nunique','Quantity':'sum','TotalRevenue':'sum'})
ProductTrend
```

```python
df.groupby('ProductScore').agg({'Quantity':'mean','TotalRevenue':'mean'})
```

# Best Customer

```python
TopCustomerSpender = df.groupby('CustomerNo')[['TotalRevenue']].sum()
TopCustomerSpender['TotalRevenue'].quantile([0.25,0.75])
```

```python
TopCustomerTransactionRate = df.groupby('CustomerNo')[['TransactionNo']].nunique()
TopCustomerTransactionRate['TransactionNo'].quantile([0.25,0.75])
```

```python
TopCustomerSpender['SpendingScore'] = pd.cut(TopCustomerSpender['TotalRevenue'],bins=[0,1817,11510,np.Inf],labels=['Low','Medium','High'])
TopCustomerSpender =TopCustomerSpender.sort_values('TotalRevenue')
TopCustomerSpender
```

```python
TopCustomerTransactionRate['TransactionRateScore'] = pd.cut(TopCustomerTransactionRate['TransactionNo'],bins=[0,1,4,np.Inf],labels=['Low','Medium','High'])
TopCustomerTransactionRate =TopCustomerTransactionRate.sort_values('TransactionNo')
TopCustomerTransactionRate
```

```python
TopCustomerTransactionRate.groupby('TransactionRateScore').agg({'TransactionNo':['min','max','mean','median','count']})
```

```python
CustomerPerformance = TopCustomerTransactionRate.merge(TopCustomerSpender, on='CustomerNo',how='left')
CustomerPerformance
```

```python
CustomerItem = df.groupby('CustomerNo')[['Quantity']].sum()

CustomerPerformance['PerformanceScore'] = CustomerPerformance.apply(lambda row:f"{row['TransactionRateScore']}_{row['SpendingScore']}",axis=1)
CustomerPerformance = CustomerPerformance.merge(CustomerItem,on='CustomerNo',how='left')

CustomerPerformance
```

```python
CustomerPerformance['PerformanceScore'].value_counts()
```

```python
CustomerPerformance['PerformanceScore'] = CustomerPerformance['PerformanceScore'].map({
    'High_High':'Best Performance','High_Medium':'Good Performance',
    'High_Low':'Standard Performance','Medium_High':'Good Performance',
    'Medium_Medium':'Standard Performance','Medium_Low':'Standard Performance',
    'Low_High':'Standard Performance','Low_Medium':'Standard Performance','Low_Low':'Low Performance'})

CustomerPerformance
```

```python
CustomerPerformance['PerformanceScore'].value_counts()
```

```python
CP = CustomerPerformance.drop(columns=['TransactionNo','TotalRevenue','Quantity'])
CP
```

```python
CustomerPerformance.groupby('PerformanceScore').agg({'TransactionNo':'sum','Quantity':'sum','TotalRevenue':'sum'})
```
